### A program to find the average of numbers from 10 to 20 (both inclusive)

```python
sum = 0

num = 0

for i in range(10, 21):
    sum += i
    num += 1

print("The sum : ", sum)
print("Total numbers in the range : ", num)
print("The average : :", sum/num)
```

```python
# Program to find the sum of all numbers until the user enters -999

sum = 0

num = input("Enter a number (-999 to exit) ")
num = int(num)

while(num != -999):
    sum += num #sum = sum + num
    num = input("Enter a number (-999 to exit) ")
    num = int(num)

print("Sum of number is ", sum)
```

```python
# Program to find the average of all numbers until the user enters -999

sum = 0

items = 0

num = input("Enter a number (-999 to exit) ")
num = int(num)

while(num != -999):
    items += 1
    sum += num #sum = sum + num
    num = input("Enter a number (-999 to exit) ")
    num = int(num)
    
if(items !=0 ):
    print("Average of numbers is ", sum/items)
else:
    print("No numbers are provided, no average")
```
